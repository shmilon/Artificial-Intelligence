```python
# Load the image using the imread() function from the cv2 library
image = cv2.imread('img.jpg')

# Create a mask with the same size as the image and set all pixels to zero (black)
mask = np.zeros(image.shape[:2], np.uint8)

# Set up the grabcut algorithm
bgdModel = np.zeros((1,65), np.float64)
fgdModel = np.zeros((1,65), np.float64)

# Define the bounding box of the object you want to keep
x,y,w,h = 10,10,300,300

# Run the grabcut algorithm
cv2.grabCut(image, mask, (x,y,w,h), bgdModel, fgdModel, 5, cv2.GC_INIT_WITH_RECT)

# Create a mask of the foreground pixels
mask_foreground = np.where((mask==2)|(mask==0), 0, 1).astype('uint8')

# Multiply the image by the mask to remove the background
image_foreground = image * mask_foreground[:,:,np.newaxis]

# Display the image with the background removed using the imshow() function from matplotlib
plt.imshow(image_foreground)
plt.show()
```
